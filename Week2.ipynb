```python
img = cv.imread('/content/drive/MyDrive/PCVK/images/kitten.jpeg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

img_rb = img_rgb.copy()
img_gb = img_rgb.copy()

# RB channel: Set Green channel ke 0
img_rb[:, :, 1] = 0

# GB channel: Set Red channel ke 0
img_gb[:, :, 0] = 0

# Menampilkan Gambar Red-Blue
plt.figure(figsize=(5, 5))
plt.imshow(img_rb)
plt.title('Red-Blue')

# Menampilkan Gambar Green-Blue
plt.figure(figsize=(5, 5))
plt.imshow(img_gb)
plt.title('Green-Blue')
```

```python
import random
black_img3 = np.zeros(shape=(512,512,3), dtype=np.int16)
def draw_random_rectangle(black_img3):
    x1 = random.randint(0, black_img3.shape[1] - 1)
    y1 = random.randint(0, black_img3.shape[0] - 1)
    x2 = random.randint(x1 + 10, black_img3.shape[1])
    y2 = random.randint(y1 + 10, black_img3.shape[0])

    # Menentukan warna acak dalam format BGR
    color = [random.randint(0, 255) for _ in range(3)]

    # Menggambar kotak
    cv.rectangle(black_img3, (x1, y1), (x2, y2), color, 3)  # Jika -1 berarti diisi penuh

# Menggambar 5 kotak acak
for _ in range(5):
    draw_random_rectangle(black_img3)

# Menampilkan gambar
plt.imshow(black_img3)
plt.title('5 Random Colored Rectangles')
```

# TUGAS KELOMPOK

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

url = "https://www.beritariau.com/foto_berita/2022/05/2022-05-23-aturan-baru-buat-ktp-nama-maksimal-60-huruf-termasuk-spasi.b686f703"
image = io.imread(url)
plt.imshow(image)

center_coordinates = (400, 305)
axes_lengths = (60, 10)
cv.ellipse(image, center_coordinates, axes_lengths, 0, 0, 360, (255, 0, 0), 3)

points = np.array([[370, 270], [590, 270], [550, 290], [350, 290]], np.int32)
points = points.reshape((-1, 1, 2))
cv.polylines(image, [points], isClosed=True, color=(0, 0, 255), thickness=3)

plt.imshow(image)
```
