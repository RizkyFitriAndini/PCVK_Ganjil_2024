```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('/content/drive/MyDrive/PCVK/images/kitten.jpeg')
plt.imshow(img)
```

```python
img2 = img
img3 = cv.cvtColor(img,cv.COLOR_BGR2RGB)  #konversi channer BGR -> RGB
plt.imshow(img3)
```

```python
img_gray = cv.imread('/content/drive/MyDrive/PCVK/images/kitten.jpeg', cv.IMREAD_GRAYSCALE)
#plt.imshow(img_gray)
#plt.imshow(img_gray, cmap='gray')
plt.imshow(img_gray, cmap='magma')
```
